```python
# import necessaries libraries
import pandas as pd 
import glob
import json
import os
```

### merging spotify data

```python
# define the folder where your files are
folder_path = 'spotify_data' 

# 2. find all files ending in .json
file_pattern = os.path.join(folder_path, "*.json.json")
file_list = glob.glob(file_pattern)

all_history = []

# 3. loop through each file and add it to our list
for file in file_list:
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)
        all_history.extend(data)

# 4. convert to a table (DataFrame) for easy viewing
df = pd.DataFrame(all_history)

print(f"Loaded {len(df)} stream records.")
print(df.head())
```
